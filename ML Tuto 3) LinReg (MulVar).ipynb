```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
df.bedrooms.median()
```

```python
import math

median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)   #df.price is the target , #fit used to training set
```

```python
#age, bedroom, age = independent variables , price = dependent variables

#Independent variables (Features/Predictors)  - the variables you control or observe to see how they affect the outcome.
#Dependent variables(Target/Output) - the variable you're trying to predict or explain.
```

```python
reg.coef_
```

```python
reg.intercept_
```
